```python
import random
```

```python
class FakeLLM:
    def __init__(self):
        print('LLM Created')
    

    def predict(self, prompt):
        responses = [
            "This is a generated response.",
            "Here are some notes based on the text.",
            "Quiz questions generated from the text.",
            "Merged document of notes and quiz."
        ]
        return random.choice(responses)
```

```python
llm= FakeLLM()
llm.predict('what is AI?')
```

```python
class pseudoLLM:
    def __init__(self, template, input_var):
        self.template = template
        self.input_var = input_var
    
    def format(self, input_dict):
        return self.template.format(**input_dict)
        
```

```python
template= pseudoLLM(
    template= "Summarize the research paper titled '{paper_input}",
    input_var= ['paper_input']
    
)
```

```python
template.format({'paper_input': 'Attention Is All You Need'})
```
